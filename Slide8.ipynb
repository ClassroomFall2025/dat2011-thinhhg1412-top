```python
# Ví dụ về metacharacters
import re
# định nghĩa pattern bắt đầu bằng 'a', kết thúc bằng 's' và có đúng 3 ký tự ở giữa
# dùng ^a...s$ để biểu diễn pattern này
pattern = '^a...s$'
test_string = 'abdds'
result = re.match(pattern, test_string)
if result:
    print("Tim kiem thanh cong.")
else:
    print("Tim kiem that bai.")
```

### \A - Khớp với các ký tự theo sau nó nằm ở đầu chuỗi

```python
# ### Ứng dụng của Regular Expressions trong Python với \A - Khớp với các ký tự theo sau nó nằm ở đầu chuỗi
import re
pattern = r"\AHello"
test_string = "Hello, world!"
result = re.match(pattern, test_string)
print("tim kiem thanh cong." if result else "tim kiem that bai.")
```

### \b - Khớp với các ký tự được chỉ định nằm ở đầu hoặc cuối của

### \w - Khớp với bất kỳ ký tự chữ cái và chữ số nào, tương đương

### Kiểm tra định dạng mật khẩu mạnh. Yêu cầu: Viết hàm trong Python để kiểm tra xem một chuỗi
### định dạng mật khẩu có mạnh hay không. Mật khẩu mạnh cần có ít nhất 8 ký tự, bao gồm chữ hoa, chữ thường, số
### và ký tự đặc biệt.

```python
# kiểm tra mật khẩu có mạnh hay không?
import re

def check_password(mat_khau):
    if len(mat_khau) < 8:
        return f"mật khẩu:{mat_khau} không đủ 8 ký tự"
    if not re.search(r"[A-Z]",mat_khau):
        return f"{mat_khau} không chứa ký tự in Hoa"
    if not re.search(r"[a-z]",mat_khau):
        return f"{mat_khau} không chứa ký tự thường"
    if not re.search(r"[0-9]",mat_khau):
        return f"{mat_khau} không chứa ký tự số"
    ky_tu_dac_biet = r'[@!#$%^&()_+?{}?]'
    if not re.search(ky_tu_dac_biet,mat_khau):
        return f"{mat_khau} không chứa ký tự đặc biệt"
    return f"mật khẩu: {mat_khau} mạnh" 
print(check_password("abc123"))
print(check_password("Abc@134r"))
```
