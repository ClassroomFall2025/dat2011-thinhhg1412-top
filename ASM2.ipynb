```python
import random

# ==============================
# Lá»šP CHA: NHÃ‚N VIÃŠN
# ==============================
class Nhanvien:
    def __init__(self, id, hoten, luong):
        self.id = id
        self.hoten = hoten
        self.luong = float(luong)

    def tong_luong(self):
        return self.luong

    def __str__(self):
        return f'ID: {self.id}, Há» tÃªn: {self.hoten}, LÆ°Æ¡ng: {self.luong:.0f}'


# ==============================
# Lá»šP CON: HÃ€NH CHÃNH
# ==============================
class Hanhchinh(Nhanvien):
    def tinh_thue(self):
        if self.luong <= 9000000:
            return 0
        elif self.luong <= 15000000:
            return (self.luong - 9000000) * 0.01
        else:
            return (self.luong - 15000000) * 0.12

    def tong_luong(self):
        return self.luong - self.tinh_thue()


# ==============================
# Lá»šP CON: TIáº¾P THá»Š
# ==============================
class Tiepthi(Hanhchinh):
    def __init__(self, id, hoten, luong, doanhso, hoahong=0.05):
        super().__init__(id, hoten, luong)
        self.doanhso = doanhso
        self.hoahong = hoahong

    def tong_luong(self):
        tong_thunhap = self.luong + (self.doanhso * self.hoahong)
        return tong_thunhap - self.tinh_thue()


# ==============================
# Lá»šP CON: TRÆ¯á»NG PHÃ’NG
# ==============================
class TruongPhong(Hanhchinh):
    def __init__(self, id, hoten, luong, luong_trachnhiem):
        super().__init__(id, hoten, luong)
        self.luong_trachnhiem = luong_trachnhiem

    def tong_luong(self):
        tong_thunhap = self.luong + self.luong_trachnhiem
        return tong_thunhap - self.tinh_thue()


# ==============================
# HÃ€M Tá»° Äá»˜NG Táº O NHÃ‚N VIÃŠN
# ==============================
def tao_ngau_nhien_nhanvien(so_luong=5):
    ho_ten_mau = ["Nguyen Van A", "Le Thi B", "Tran Van C", "Pham Thi D", "Vo Van E"]
    nhanvien_list = []

    for i in range(so_luong):
        loai = random.choice(["hanhchinh", "tiepthi", "truongphong"])
        ma = f"NV{i+1:03d}"
        hoten = random.choice(ho_ten_mau)
        luong = random.randint(8000000, 20000000)

        if loai == "hanhchinh":
            nv = Hanhchinh(ma, hoten, luong)
        elif loai == "tiepthi":
            doanhso = random.randint(10000000, 100000000)
            hoahong = round(random.uniform(0.03, 0.07), 2)
            nv = Tiepthi(ma, hoten, luong, doanhso, hoahong)
        else:
            luong_trachnhiem = random.randint(2000000, 5000000)
            nv = TruongPhong(ma, hoten, luong, luong_trachnhiem)

        nhanvien_list.append(nv)

    return nhanvien_list


# ==============================
# HÃ€M NHáº¬P NHÃ‚N VIÃŠN Tá»ª BÃ€N PHÃM
# ==============================
def nhap_nhanvien():
    nhanvien_list = []
    while True:
        print("\n-- Nháº­p thÃ´ng tin nhÃ¢n viÃªn --")
        id = input("MÃ£ nhÃ¢n viÃªn: ")
        hoten = input("Há» tÃªn: ")
        luong = float(input("LÆ°Æ¡ng cÆ¡ báº£n: "))

        print("Chá»n loáº¡i nhÃ¢n viÃªn:")
        print("1. HÃ nh chÃ­nh")
        print("2. Tiáº¿p thá»‹")
        print("3. TrÆ°á»Ÿng phÃ²ng")
        loai = input("Chá»n (1-3): ")

        if loai == "1":
            nv = Hanhchinh(id, hoten, luong)
        elif loai == "2":
            doanhso = float(input("Nháº­p doanh sá»‘: "))
            hoahong = float(input("Nháº­p hoa há»“ng (vd 0.05 = 5%): "))
            nv = Tiepthi(id, hoten, luong, doanhso, hoahong)
        elif loai == "3":
            luong_trachnhiem = float(input("Nháº­p lÆ°Æ¡ng trÃ¡ch nhiá»‡m: "))
            nv = TruongPhong(id, hoten, luong, luong_trachnhiem)
        else:
            print("âŒ Loáº¡i khÃ´ng há»£p lá»‡.")
            continue

        nhanvien_list.append(nv)
        tiep = input("Báº¡n cÃ³ muá»‘n nháº­p thÃªm (y/n)? ").lower()
        if tiep != "y":
            break
    return nhanvien_list


# ==============================
# GHI & Äá»ŒC FILE
# ==============================
def ghi_file(nhanvien_list, ten_file="Nhanvien.txt"):
    with open(ten_file, "w", encoding="utf-8") as f:
        for nv in nhanvien_list:
            f.write(str(nv) + "\n")
    print(f"âœ… ÄÃ£ ghi {len(nhanvien_list)} nhÃ¢n viÃªn vÃ o file '{ten_file}'")


def doc_file(ten_file="Nhanvien.txt"):
    print("\nğŸ“„ Danh sÃ¡ch nhÃ¢n viÃªn trong file:")
    with open(ten_file, "r", encoding="utf-8") as f:
        print(f.read())


# ==============================
# MENU CHÃNH
# ==============================
def menu():
    ds = []
    while True:
        print("\n===== QUáº¢N LÃ NHÃ‚N VIÃŠN =====")
        print("1. Nháº­p nhÃ¢n viÃªn báº±ng tay")
        print("2. Táº¡o ngáº«u nhiÃªn danh sÃ¡ch nhÃ¢n viÃªn")
        print("3. Ghi danh sÃ¡ch ra file")
        print("4. Äá»c danh sÃ¡ch tá»« file")
        print("0. ThoÃ¡t")

        chon = input("Chá»n chá»©c nÄƒng: ")
        if chon == "1":
            ds = nhap_nhanvien()
        elif chon == "2":
            so_luong = int(input("Nháº­p sá»‘ lÆ°á»£ng nhÃ¢n viÃªn muá»‘n táº¡o: "))
            ds = tao_ngau_nhien_nhanvien(so_luong)
            print(f"âœ… ÄÃ£ táº¡o {so_luong} nhÃ¢n viÃªn ngáº«u nhiÃªn.")
        elif chon == "3":
            if ds:
                ghi_file(ds)
            else:
                print("âš ï¸ ChÆ°a cÃ³ danh sÃ¡ch nhÃ¢n viÃªn Ä‘á»ƒ ghi.")
        elif chon == "4":
            doc_file()
        elif chon == "0":
            print("ğŸ‘‹ ThoÃ¡t chÆ°Æ¡ng trÃ¬nh.")
            break
        else:
            print("âŒ Lá»±a chá»n khÃ´ng há»£p lá»‡.")


# ==============================
# CHáº Y CHÆ¯Æ NG TRÃŒNH
# ==============================
if __name__ == "__main__":
    menu()
```
