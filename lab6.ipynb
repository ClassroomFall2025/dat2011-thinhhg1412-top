## Lab 6 - Bài 1

```python
# Code lab 6 bài 1 ở đây
class ChuNhat:
    def __init__(self, dai, rong):
        self.dai = dai
        self.rong = rong

    def get_chu_vi(self):
        return (self.dai + self.rong) * 2

    def get_dien_tich(self):
        return self.dai * self.rong

    def xuat(self):
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        print(f"Diện tích: {self.get_dien_tich()}")
        print(f"Chu vi: {self.get_chu_vi()}")


class Vuong(ChuNhat):
    def __init__(self, canh):
        super().__init__(canh, canh)
        self.canh = canh

    def xuat(self):
        print(f"Cạnh: {self.canh}")
        print(f"Diện tích: {self.get_dien_tich()}")
        print(f"Chu vi: {self.get_chu_vi()}")


# Chương trình chính
print("Nhập hình chữ nhật thứ 1:")
dai1 = float(input("Nhập chiều dài: "))
rong1 = float(input("Nhập chiều rộng: "))
cn1 = ChuNhat(dai1, rong1)

print("\nNhập hình chữ nhật thứ 2:")
dai2 = float(input("Nhập chiều dài: "))
rong2 = float(input("Nhập chiều rộng: "))
cn2 = ChuNhat(dai2, rong2)

print("\nNhập hình vuông:")
canh = float(input("Nhập cạnh: "))
vu = Vuong(canh)

print("\n--- Thông tin hình chữ nhật 1 ---")
cn1.xuat()
print("\n--- Thông tin hình chữ nhật 2 ---")
cn2.xuat()
print("\n--- Thông tin hình vuông ---")
vu.xuat()

```

## Lab 6 - Bài 2

```python
# Code lab 6  bài 2 ở đây
from abc import ABC, abstractmethod

class SinhVienPoly(ABC):
    def __init__(self, ho_ten, nganh):
        self.ho_ten = ho_ten
        self.nganh = nganh

    @abstractmethod
    def get_diem(self):
        pass

    def get_hoc_luc(self):
        diem = self.get_diem()
        if diem < 5:
            return "Yếu"
        elif diem < 7:
            return "Trung bình"
        elif diem < 8:
            return "Khá"
        elif diem < 9:
            return "Giỏi"
        else:
            return "Xuất sắc"

    def xuat(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Ngành: {self.nganh}")
        print(f"Điểm: {self.get_diem()}")
        print(f"Học lực: {self.get_hoc_luc()}")


# Lớp con kế thừa
class SinhVienIT(SinhVienPoly):
    def __init__(self, ho_ten, nganh, diem_java, diem_html, diem_css):
        super().__init__(ho_ten, nganh)
        self.diem_java = diem_java
        self.diem_html = diem_html
        self.diem_css = diem_css

    def get_diem(self):
        return (self.diem_java * 2 + self.diem_html + self.diem_css) / 4


# Chạy thử
sv = SinhVienIT("Hoàng Lê Trung Thịnh", "CNTT", 8, 7, 9)
sv.xuat()
```

## Lab 6 - Bài 3

```python
# Code lab 6  bài 3 ở đây
from abc import ABC, abstractmethod

# ===== LỚP CHA =====
class SinhVienPoly(ABC):
    def __init__(self, ho_ten, nganh):
        self.ho_ten = ho_ten
        self.nganh = nganh

    # Phương thức trừu tượng (chưa có nội dung)
    @abstractmethod
    def get_diem(self):
        pass

    # Phương thức xếp loại học lực
    def get_hoc_luc(self):
        diem = self.get_diem()
        if diem < 5:
            return "Yếu"
        elif diem < 7:
            return "Trung bình"
        elif diem < 8:
            return "Khá"
        elif diem < 9:
            return "Giỏi"
        else:
            return "Xuất sắc"

    # Phương thức xuất thông tin
    def xuat(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Ngành: {self.nganh}")
        print(f"Điểm: {self.get_diem():.2f}")
        print(f"Học lực: {self.get_hoc_luc()}")


# ===== LỚP CON 1: SINH VIÊN IT =====
class SinhVienIT(SinhVienPoly):
    def __init__(self, ho_ten, nganh, diem_java, diem_html, diem_css):
        super().__init__(ho_ten, nganh)
        self.diem_java = diem_java
        self.diem_html = diem_html
        self.diem_css = diem_css

    # Ghi đè phương thức get_diem()
    def get_diem(self):
        return (2 * self.diem_java + self.diem_html + self.diem_css) / 4


# ===== LỚP CON 2: SINH VIÊN BIZ =====
class SinhVienBiz(SinhVienPoly):
    def __init__(self, ho_ten, nganh, diem_marketing, diem_sales):
        super().__init__(ho_ten, nganh)
        self.diem_marketing = diem_marketing
        self.diem_sales = diem_sales

    # Ghi đè phương thức get_diem()
    def get_diem(self):
        return (2 * self.diem_marketing + self.diem_sales) / 3


# ===== CHƯƠNG TRÌNH CHÍNH =====
if __name__ == "__main__":
    # Tạo sinh viên IT
    sv_it = SinhVienIT("Hoàng Lê Trung Thịnh", "Công nghệ thông tin", 8, 7, 9)

    # Tạo sinh viên Biz
    sv_biz = SinhVienBiz("Trần Văn Tèo", "Kinh doanh", 7, 8)

    # Xuất thông tin
    print("=== SINH VIÊN IT ===")
    sv_it.xuat()

    print("\n=== SINH VIÊN BIZ ===")
    sv_biz.xuat()
```

## Lab 6 - Bài 4

```python
# code lab 6 bài 4 ở đây
class SinhVienPoly:
    def __init__(self, ho_ten, nganh):
        self.ho_ten = ho_ten
        self.nganh = nganh

    def get_diem(self): pass

    def get_hoc_luc(self):
        d = self.get_diem()
        if d < 5: return "Yếu"
        elif d < 7: return "Trung bình"
        elif d < 8: return "Khá"
        elif d < 9: return "Giỏi"
        else: return "Xuất sắc"

    def xuat(self):
        print(f"Họ tên: {self.ho_ten}, Ngành: {self.nganh}, Điểm: {self.get_diem():.2f}, Học lực: {self.get_hoc_luc()}")


class SinhVienIT(SinhVienPoly):
    def __init__(self, ho_ten, java, html, css):
        super().__init__(ho_ten, "IT")
        self.java, self.html, self.css = java, html, css

    def get_diem(self):
        return (2 * self.java + self.html + self.css) / 4


class SinhVienBiz(SinhVienPoly):
    def __init__(self, ho_ten, marketing, sales):
        super().__init__(ho_ten, "Biz")
        self.marketing, self.sales = marketing, sales

    def get_diem(self):
        return (2 * self.marketing + self.sales) / 3


# === CHƯƠNG TRÌNH CHÍNH ===
ds = []

while True:
    print("\n1. Nhập SV  2. Xuất DS  3. SV giỏi  4. Sắp xếp  5. Thoát")
    chon = input("Chọn: ")
    if chon == "1":
        loai = input("Ngành (IT/Biz): ").lower()
        ho_ten = input("Họ tên: ")
        if loai == "it":
            java = float(input("Điểm Java: "))
            html = float(input("Điểm HTML: "))
            css = float(input("Điểm CSS: "))
            ds.append(SinhVienIT(ho_ten, java, html, css))
        else:
            marketing = float(input("Điểm Marketing: "))
            sales = float(input("Điểm Sales: "))
            ds.append(SinhVienBiz(ho_ten, marketing, sales))
    elif chon == "2":
        for sv in ds: sv.xuat()
    elif chon == "3":
        for sv in ds:
            if sv.get_hoc_luc() == "Giỏi": sv.xuat()
    elif chon == "4":
        ds.sort(key=lambda sv: sv.get_diem(), reverse=True)
        print("Đã sắp xếp theo điểm!")
    elif chon == "5":
        break
    else:
        print("Vui lòng chọn lại!")
```
