```python
def nhap_nhan_vien():
    print(">>> Chức năng: Nhập danh sách nhân viên (Y1)")

def xuat_danh_sach():
    print(">>> Chức năng: Xuất danh sách nhân viên (Y2)")

def tim_nhan_vien_ma():
    print(">>> Chức năng: Tìm nhân viên theo mã (Y3)")

def xoa_nhan_vien():
    print(">>> Chức năng: Xóa nhân viên theo mã (Y4)")

def cap_nhat_nhan_vien():
    print(">>> Chức năng: Cập nhật thông tin nhân viên (Y5)")

def tim_theo_khoang_luong():
    print(">>> Chức năng: Tìm nhân viên theo khoảng lương (Y6)")

def sap_xep_theo_ten():
    print(">>> Chức năng: Sắp xếp nhân viên theo họ và tên (Y7)")

def sap_xep_theo_thu_nhap():
    print(">>> Chức năng: Sắp xếp nhân viên theo thu nhập (Y8)")

def top_5_thu_nhap():
    print(">>> Chức năng: Xuất 5 nhân viên có thu nhập cao nhất (Y9)")

def thoat():
    print(">>> Kết thúc chương trình. Tạm biệt!")

def menu():
    while True:
        print("\n===== MENU QUẢN LÝ NHÂN SỰ TIỀN LƯƠNG =====")
        print("1. Nhập danh sách nhân viên (Y1)")
        print("2. Xuất danh sách nhân viên (Y2)")
        print("3. Tìm nhân viên theo mã (Y3)")
        print("4. Xóa nhân viên theo mã (Y4)")
        print("5. Cập nhật thông tin nhân viên (Y5)")
        print("6. Tìm nhân viên theo khoảng lương (Y6)")
        print("7. Sắp xếp nhân viên theo họ và tên (Y7)")
        print("8. Sắp xếp nhân viên theo thu nhập (Y8)")
        print("9. Xuất 5 nhân viên có thu nhập cao nhất (Y9)")
        print("0. Thoát")
        
        chon = input("Mời bạn chọn (0-9): ")

        if chon == "1":
            nhap_nhan_vien()
        elif chon == "2":
            xuat_danh_sach()
        elif chon == "3":
            tim_nhan_vien_ma()
        elif chon == "4":
            xoa_nhan_vien()
        elif chon == "5":
            cap_nhat_nhan_vien()
        elif chon == "6":
            tim_theo_khoang_luong()
        elif chon == "7":
            sap_xep_theo_ten()
        elif chon == "8":
            sap_xep_theo_thu_nhap()
        elif chon == "9":
            top_5_thu_nhap()
        elif chon == "0":
            thoat()
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại.")

# Gọi chạy menu
menu()
```
